Script:
* The first pandas function that we will use will be the read_csv function.
* This function will load a file in the comma separated value format, or CSV for short, and return the data in an object called a data frame.
* Let's look at an example.

```python
# source: https://www.ers.usda.gov/data-products/food-price-outlook

data = pd.read_csv("https://ers.usda.gov/sites/default/files/_laserfiche/DataFiles/50673/historicalcpi.csv?v=86136")
```

Script:
* This example loads an online data set from the US Department of Agriculture looking at consumer price data over the last 30 years or so.
* Let's take a look at that data.

```python
data
```

Script:
* This data frame has 1100 rows and 3 columns of data.
* Each row has its own number or index, and there are columns for which consumer price index the row matches, the year of the data, and the Percentage change from the previous year.
* Data frames are conveniently labeled like this, and those labels distinguish structured data like this from unlabeled or unstructured data with many unidentified numbers.
* In mod 1, you will be using data frames throughout the semester since they are a big convenience.
* In mod 2, you will start using data frames about halfway through after covering Python basics and simpler data structures used to build data frames.
* Let's take a closer look at that data.
* First, let's reshape it into a more useful form.


```python
data2 = pd.pivot(data,
                 columns="Consumer Price Index item",
                 index="Year",
                 values="Percent change")
```

```python
data2
```

Script:
* If you have used pivot tables in Microsoft Excel or Google Sheets, the pivot method is very similar.
* That data looks a little more friendly, but now we can see that there are a lot of different consumer price indices being reported.
* If we plot all of those at once, it will be a mess.
* Let's pick a smaller subset to review.

```python
data2.columns
```

Script:
* I will pick "All food", "Meats", and "Fresh fruits".

```python
data3 = data2[["All food", "Meats", "Fresh fruits"]]
```

Script:
* That syntex made a new data frame with just those three columns.
* You will see many examples of Python code in mod 1 that you can adapt to your homework or projects, and mod 2 will explain the syntax more systematically.
* Let's look at the filtered data.

```python
data3
```

Script:
* Let's plot that data now.

```python
data3.plot()
```

Script:
* We can see that meats and fruits tend to have their prices go up the same years, but the overall food changes are more steady.
* Why is that?
