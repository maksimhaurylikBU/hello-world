# Video: Building a Linear Model

In this video, you will learn how to quickly build a linear model and make predictions with it.

Script:
* In this video, I will show you a simple linear model with one input variable.
* We will start by importing a few packages to build and plot the model.
* Some of these will be new, but you will learn these early in this program.

```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn.linear_model
```

Script:
* Now let's make some dummy data to model.

```python
X = [1, 2, 3, 4, 5]
```

```python
X
```

```python
Y = [2, 4, 3, 4, 5]
```

```python
Y
```

Script:
* Now I will plot it so you can see the data.

```python
plt.scatter(X, Y);
```

Script:
* Four out of those five points roughly fall on a line, but the second one is farther off.

Script:
* The scikit-learn library gives us code for lienar regression.

```python
m = sklearn.linear_model.LinearRegression()
```

```python
m
```

Script:
* To fit the model, we will need to make some tweaks to how we store our X values because scikit-learn was designed to support multiple input variables at once.
* So we will convert our list of X values into a NumPy array and reshape it into two dimensions.
* If that feels like technical mumbo jumbo, do not worry.
* Those details will be explained in a later module.

```python
X = np.array(X).reshape(-1, 1)
```

```python
X
```

Script:
* See how $X$ is now an array and displayed as one column of numbers?
* Now we can fit the model.

```python
m.fit(X, Y)
```

Script:
* Let's plot the data again, and compare the data to the model predictions.

```python
plt.scatter(X, Y, label="data")
plt.plot(X, m.predict(X), label="model")
plt.legend();
```

Script:
* Now you have seen how to build a linear model with the scikit-learn library.
* Essentially the same code will work for any number of input variables.
* I glossed over many of the details this time, but we will cover them in the next few modules.
